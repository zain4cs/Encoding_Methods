___________________________

**(One Hot Encoder) Using sklearn**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:4], df.iloc[:,-1],
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
X_test.head().reset_index()
```

________________________

**Apply Encoder**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32)
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
X_test_new = ohe.transform(X_test[['fuel','owner']])
```

```python
np.hstack((X_train[['km_driven','km_driven']].values,X_train_new)).shape
```

**(One Hot Encoder) with top Categories**

```python
count = df['brand'].value_counts()
```

```python
threshold=100
```

```python
rep = count[count <= threshold].index
```

```python
df['brand'] = df['brand'].replace(rep,'Uncommon')
pd.get_dummies(df['brand'], dtype='int')
```
